```python
import glob, re
import pandas as pd
from tabulate import tabulate


USdataJan2022 = glob.glob("./COVID-19/csse_covid_19_data/csse_covid_19_daily_reports_us/*?2022.csv")
CAdataJan2022 = [['Date', 'Daily cases']] + [[0, 0]] * 30
currentDay = 0

for day in USdataJan2022:
    dom = int(str(day[-11 : -9]).lstrip("0"))
    splitData = open(day).read()
    cases = int(re.findall("California,[^,]*,[^,]*,[^,]*,[^,]*,([^,]*)", splitData)[0])
    CAdataJan2022[dom] = ["Jan " + str(dom) + " 2022", cases]
    currentDay = max(dom, currentDay) 



peakDay, i = (1, 0), 2
daily = [['Date', 'Daily cases']] + [[0, 0]] * 30
while i < currentDay:
    delta = CAdataJan2022[i][1] - CAdataJan2022[i-1][1]
    daily[i] = [CAdataJan2022[i][0], delta]
    if delta > peakDay[1]:
        peakDay = (i, delta)
    i+=1
print(tabulate(daily[2:currentDay+1]))
print("Jan " + str(peakDay[0]) + " had the most new cases this month.")
```
